**1.1. CHO MỘT CHUỖI MẪU NHƯ SAU: chuoi_mau = "Hôm nay công ty bán được 123 đơn hàng, hôm sau sẽ bán được 456 hoặc 789 đơn hàng. Không 1 đơn nào bị hủy cả. 1,2,3 cùng cố gắng nhé! ". VIẾT CHƯƠNG TRÌNH LỌC RA CÁC SỐ NGUYÊN DƯƠNG TỪ 3 CHỮ SỐ TRỞ LÊN.**

```python
import re

# Chuỗi mẫu
chuoi_mau = "Hôm nay công ty bán được 123 đơn hàng, hôm sau sẽ bán được 456 hoặc 789 đơn hàng. Không 1 đơn nào bị hủy cả. 1,2,3 cùng cố gắng nhé!"

# Biểu thức chính quy để lọc ra các số nguyên dương từ 3 chữ số trở lên
pattern = r'\b\d{3,}\b'

# Lọc ra các số nguyên dương từ 3 chữ số trở lên từ chuỗi mẫu
numbers = re.findall(pattern, chuoi_mau)

# In kết quả
print(chuoi_mau)
print("Các số nguyên dương có ít nhất 3 chữ số", numbers)
```

### **CÂU 2**

**YÊU CẦU:**

**VIẾT CHƯƠNG TRÌNH TÍNH TIỀN TAXI. CHO NGƯỜI DÙNG NHẬP VÀO SỐ KM ĐÃ ĐI, CÓ THẺ THÀNH VIÊN HAY KHÔNG, CÓ MÃ GIẢM GIÁ HAY KHÔNG. GIẢ SỬ CHO GIÁ TAXI CỐ ĐỊNH LÀ 12,00/KM. NẾU CÓ THẺ THÀNH VIÊN THÌ GIẢM 10% TỔNG TIỀN, CÓ MÃ GIẢM GIÁ THÌ GIẢM THÊM 10% TỔNG TIỀN NỮA.**

```python
def tinh_tien_taxi(km, co_the_thanh_vien, co_ma_giam_gia):
    gia_taxi = 12000  # Giá taxi cố định là 12,000/Km
    tong_tien = gia_taxi * km  # Tổng tiền trước khi giảm giá

    # Kiểm tra nếu có thẻ thành viên, giảm 10% tổng tiền
    if co_the_thanh_vien:
        tong_tien -= tong_tien * 0.1

    # Kiểm tra nếu có mã giảm giá, giảm thêm 10% tổng tiền
    if co_ma_giam_gia:
        tong_tien -= tong_tien * 0.1

    return tong_tien

# Nhập số Km đã đi
km = float(input("Nhập số Km đã đi: "))

# Nhập thông tin về thẻ thành viên và mã giảm giá
co_the_thanh_vien = input("Có thẻ thành viên không? (Có/Không): ").strip().lower() == 'có'
co_ma_giam_gia = input("Có mã giảm giá không? (Có/Không): ").strip().lower() == 'có'

# Tính tiền taxi và in ra kết quả
tong_tien = tinh_tien_taxi(km, co_the_thanh_vien, co_ma_giam_gia)
print(f"Tổng tiền taxi là: {tong_tien:,} VND")
```

**3.6. CÓ SỰ KHÁC BIỆT TRONG TIÊU THỤ ĐIỆN GIỮA CÁC LOẠI NGÀY (WEEKDAY VÀ WEEKEND) KHÔNG (TÍNH MỨC ĐIỆN TIÊU THỤ TRUNG BÌNH THEO NGÀY CHO 5 NGÀY TRONG TUẦN VÀ 2 NGÀY CUỐI TUẦN, SAU ĐÓ THỰC HIỆN SO SÁNH 2 GIÁ TRỊ NÀY)? VẼ BIỂU ĐỒ MINH HỌA VÀ NHẬN XÉT.**

```python
# Tạo một cột mới 'Day_Type' để đánh dấu loại ngày (weekday hoặc weekend)
SteelData['Day_Type'] = SteelData['date'].dt.dayofweek.apply(lambda x: 'Weekday' if x < 5 else 'Weekend')

# Tính mức tiêu thụ điện trung bình theo loại ngày (weekday và weekend)
usage_by_day_type = SteelData.groupby('Day_Type')['Usage_kWh'].mean()

# Vẽ biểu đồ minh họa
usage_by_day_type.plot(kind='bar', color=['blue', 'red'])
plt.title('Total Energy Consumption by Week Status')
plt.xlabel('Week Status')
plt.ylabel('Total Energy Consumption (kWh)')
plt.xticks(rotation=0)
plt.grid(axis = 'y', linestyle = '-', alpha=0.7)
plt.grid(axis = 'x', linestyle = '-', alpha = 0.7)
plt.tight_layout()
```

**NHẬN XÉT:**

Biểu đồ cột so sánh tổng điện năng tiêu thụ các các công ty sản xuất thép giữa các ngày trong tuần và 2 ngày cuối tuần.

Nhìn chung, có thể thấy được mức điện năng tiêu thụ điện các ngày trong tuần nhiều hơn 2 ngày cuối tuần. Cụ thể:
+ Tổng điện năng tiêu thụ điện các ngày trong tuần là khoảng 33 kWh
+ Tổng điện năng tiêu thụ điện 2 ngày cuối tuần khoảng 12 kWh.
+ Mức chênh lệch khoảng 21 kWh


```python
# Tính mức điện tiêu thụ trung bình theo ngày cho cả các ngày trong tuần và các ngày cuối tuần
usage_by_day = SteelData.groupby('Day_of_week')['Usage_kWh'].sum()

# Vẽ biểu đồ minh họa
usage_by_day.plot(kind='bar', color=['blue', 'blue', 'blue', 'blue', 'blue', 'red', 'red'])  # Các ngày trong tuần màu xanh, các ngày cuối tuần màu đỏ
plt.title('Total Energy Consumption by Week Status')
plt.xlabel('Weekday & Weekend')
plt.ylabel('Total Energy Consumption (kWh)')
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], rotation=45)
plt.grid(axis = 'y', linestyle = '-', alpha = 0.7)
plt.grid(axis = 'x', linestyle = '-', alpha = 0.7)
plt.tight_layout()
```

**NHẬN XÉT:**

Qua biểu đồ cột chi tiết, có thể thấy được mức tiêu thụ điện năng của các ngày cuối tuần cũng đạt mức cao, từ đó có thể kết luận rằng không có sự phụ thuộc nào về mức tiêu thụ điện năng giữa các ngày trong tuần và 2 ngày cuối tuần trong ngành công nghiệp sản xuất thép.

```python
CityPowerConsumption['Total Power Consumption'] = CityPowerConsumption['Zone 1 Power Consumption'] + CityPowerConsumption['Zone 2 Power Consumption'] + CityPowerConsumption['Zone 3 Power Consumption']
CityPowerConsumption
```

**_B. Vẽ biểu đồ_**

```python
CityPowerConsumption.plot.scatter(x = 'Humidity', y = 'Total Power Consumption', figsize = (10, 8), s = 5, alpha = 0.5)
plt.title('Relationship between Humidity vs Power Consumption')
plt.xlabel('Humidity')
plt.ylabel('Power Consumption')
plt.grid()
```

**NHẬN XÉT:**

Biểu đồ Scatter (phân tán) cho thấy độ tập trung của mức độ sử dụng điện năng tại thành phố dựa theo Humidity (độ ẩm) tập trung chủ yếu ở mức độ ẩm từ 40 đến 90 và mức điện năng tiêu thụ trong khoảng 40 000 kWh đến hơn 100 000 kWh.

**_B.1. Tạo các nhóm cho dữ liệu Humidity_**

```python
bins = [0, 20, 40, 60, 80, 100]
labels = ['0-20', '20-40', '40-60', '60-80', '80-100']
CityPowerConsumption['Humidity_Group'] = pd.cut(CityPowerConsumption['Humidity'], bins = bins, labels = labels, right = False)
CityPowerConsumption
```

**_B.2. Tính toán MEAN cho mỗi nhóm_**

```python
PowerConsumptionByHumidity = CityPowerConsumption.groupby('Humidity_Group')['Total Power Consumption'].mean()
PowerConsumptionByHumidity
```

**_B.3. Vẽ biểu đồ bar_**

```python
PowerConsumptionByHumidity.plot.bar(legend = True)
plt.title('Power Consumption versus Humidity (tính theo mean Total Power Consumption)')
plt.ylabel('Power Consumption')
plt.xlabel('Humidity_Bin')
plt.grid(True, linestyle = '--', alpha = 0.3);
```

**NHẬN XÉT:**

Biểu đồ cột cho thấy được mức tiêu thụ điện năng phân bổ theo độ ẩm. Cụ thể:
+ Tại mức độ ẩm (0 đến 20): Mức độ sử dụng điện là nhiều nhất (~ 90 000 kWh)
+ Tại mức độ ẩm (80 đến 100): Mức độ sử dụng điện là thấp nhất (~ 65 000 kWh)

Biểu đồ còn cho thấy: 
+ Mức độ sử dụng điện **tỷ lệ nghịch** với độ ẩm (độ ẩm trong khoảng 0 đến 100)

**NHẬN XÉT:**

Biểu đồ phân tán trên cho ta thấy được độ tập trung của mức độ sử dụng điện năng tại thành phố dựa theo Temperature (Nhiệt độ) tập trung chủ yếu ở mức nhiệt độ từ 10 đến 25 độ và mức điện năng tiêu thụ trong khoảng 40 000 kWh đến hơn 100 000 kWh.

Từ đó thấy được, so với đồ ẩm (Humidity) thì nhiệt độ là yếu tố khiến cho mức độ tiêu thụ điện năng của thành phó TETUAN tăng cao!

**_B.1. Tạo các nhóm cho dữ liệu Temperature_**

```python
CityPowerComsumptionByTemperature.plot.bar(legend = True)
plt.title('Power Consumption versus Temperature (tính theo mean Total Power Consumption)')
plt.ylabel('Power Consumption')
plt.xlabel('Temperature_Bin')
plt.grid(True, linestyle = '--', alpha = 0.3);
```

**NHẬN XÉT:**

Biểu đồ cột cho thấy có sự **tương quan thuận chiều** giữa mức độ sử dụng điện và nhiệt độ (khoảng nhiệt độ từ 0 - 35 độ), cụ thể:
+ Từ 0 đến 5 độ ghi nhận mức tiêu thụ điện năng thấp nhất (~ 50 000 kWh)
+ Từ 30 đến 35 độ ghi nhận mức tiêu thụ điện năng cao nhất (~ 95 000 kWh)
